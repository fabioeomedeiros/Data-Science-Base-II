```python
# Importando a biblioteca pandas
import pandas as pd
# Importando a base de dados
tabela = pd.read_csv('datas/telecom_users.csv')
# Visualisar a base de dados
display(tabela)
```

```python
print(tabela['Genero'].value_counts())
print((tabela['Genero'].value_counts() / len(tabela['Genero'])) * 100)
```

```python
print(tabela['ServicoInternet'].value_counts())
print((tabela['ServicoInternet'].value_counts() / len(tabela['ServicoInternet'])) * 100)
```

```python
print(tabela['TipoContrato'].value_counts())
print((tabela['TipoContrato'].value_counts() / len(tabela['TipoContrato'])) * 100)
```

```python
print(tabela['FormaPagamento'].value_counts())
print((tabela['FormaPagamento'].value_counts() / len(tabela['FormaPagamento'])) * 100)
```

```python
print(tabela['Churn'].value_counts())
print((tabela['Churn'].value_counts() / len(tabela['Churn'])) * 100)
```
